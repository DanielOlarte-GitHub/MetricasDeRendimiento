## 1. Limpieza de valores NaN

### 1.1 Listar la cantidad de valores NaN por columna

```python
#listar valores vacios
df.isnull().sum()
```

### 1.2.- Eliminar los valores NaN del dataset

```python
#eliminar los valores vacios
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#separar mes y año del df
df['year'] = pd.DatetimeIndex(df['join_date']).year
df['month'] = pd.DatetimeIndex(df['join_date']).month
```

```python
#Se imprime informacion por cada columna del df
df.info()
```

## 2. ¿Cuántas reuniones por grupo se registran en el dataset?

```python
#numero de registros que tiene cada grupo
#OPCION1 Manual:

print("houston machine learning: ",df[df['group']=='houston machine learning'].shape)
print("houstonr: ",df[df['group']=='houstonr'].shape)
print("houston data science: ",df[df['group']=='houston data science'].shape)
print("houston energy data science: ",df[df['group']=='houston data visualization '].shape)
print("houston data visualization: ",df[df['group']=='houston energy data science '].shape)

        
```

```python
#numero de registros que tiene cada grupo
#OPCION2: Metodo
df['group'].value_counts()
```

<h5><b>En el dataset se registran las reuniones por grupo de la siguiente manera:

en Houston data science se registran 2150 reuniones, en Houston energy data science se registran 1891 reuniones, seguido de Houston data visualization con 1709 reuniones, Houstonr con 1037 reuniones y por ultimo Houston machine learning con 740 reuniones

## 3. Hacer un gráfico (Pastel) de reuniones por grupo en el dataset (similar o mejor al modelo ejemplo)

```python
#hacer grafico de pastel por cada grupo
df['group'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(10,10))
plt.title('Grupos') #Se asigna el titulo
plt.show() #Se grafica
```
